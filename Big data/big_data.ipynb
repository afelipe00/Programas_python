```python
import pandas as pd
```

```python
df_price = pd.read_csv("price_paid_records.csv")
df_price
```

```python
#mapper 1
sr_date =  df_price['Date of Transfer']
lst_year = []
for date in sr_date:
    year = date.split('-')
    lst_year.append((year[0],1))
lst_year
```

```python
#ordenar
lst_year.sort()
```

```python
#reducer
srs_years = pd.Series(lst_year)
srs_years.value_counts()
```

```python
#Town/City
#price
df_prom = pd.DataFrame(df_price['Town/City'])
df_prom['Price'] = df_price['Price']
df_prom
```

```python
lst_prom = []
for fill,row in df_prom.iterrows():
    lst_prom.append((row['Town/City'],row['Price']))
lst_prom
```

```python
srs_prom = pd.Series(lst_prom)
srs_prom.mean(axis=0)
```

```python
#reducer3
dic_minimo = {}
for col,row in df_prom:
    if row['Town/City'] in dic_minimo:
        if dic_minimo[row['Town/City']] > row['Price']:
            dic_minimo[row['Town/City']] = row['Price']
    else:
        dic_minimo[row['Town/City']] = row['Price']
dic_minimo
```
