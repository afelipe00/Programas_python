```python
#enumeracion exhaustiva
objetivo = int(input('escoge un entero'))
respuesta = 0
while respuesta**2 < objetivo:
    ##print(respuesta)
    respuesta+=1
if respuesta ** 2 == objetivo:
    print(f'la raiz cuadrada de {objetivo} es {respuesta}')
else:
    print('no tiene raiz cuadrada exacta')
```

```python
#aproximacion de solucion
#cuando se aproxima una solucion tenemo que definir el error (a este error que es la diferencia entre la realidad y la aprox) se le llama epsilon
#mientras mas preciso se quiere ser mas ciclos de computo se necesitan
#definir nuestro tradeoff nuestro canje
objetivo = int(input('ingrese un numero'))
epsilon = 0.01
paso = epsilon**2
respuesta = 0.0
while abs(respuesta**2 -  objetivo) >= epsilon and respuesta <= objetivo:
    print(abs(respuesta**2 -  objetivo), respuesta)
    respuesta+= paso
if abs(respuesta**2 - objetivo) >= epsilon:
    print(f'no se encontro la raiz de {objetivo}')
else:
    print(f'la raiz de {objetivo} es {respuesta}')
```

```python
#busqueda binaria
#este tipo de busqueda funciona cuando el conjunto tiene un orden
objetivo =  int(input('ingrese el numero: '))
epsilon = 0.001
bajo = 0.0
alto = max(1.0, objetivo)
respuesta = (alto + bajo)/2

while abs(respuesta**2 - objetivo) >= epsilon: #la funcion abs retorna el valor absoluto
    print(f'el bajo {bajo} alto {alto} respuesta {respuesta}')
    if respuesta**2 < objetivo:
        bajo = respuesta
    else:
        alto = respuesta
    respuesta = (alto + bajo) / 2
print(f'la raiz cuadrada de  {objetivo} es {respuesta}')
```

```python
#funciones y abstraccion
"""
las funciones de python pueden tener argumentos de tipo keyword y valores por defecto
parametros: son los valores que la funcion recibe en el encabezado def funcion (parametros)
argumentos: son los valores enviados a la funcion
keyword: palabra clave
algo especial en pytho en que las funciones tambien pueden ser enviadas como argumentos, esto se debe a que en python todos los elementos se manejan como objetos
"""
def suma (a,b):
    total = a + b
    return total
suma(5,6)
```

```python
def nombre_completo(nombre, apellido,invertido = False):
    if invertido:
        return f'{apellido}{nombre}'
    else:
        return f'{nombre} {apellido}'
#formas de llamar a una funcion
#por orden de parametros (de forma posicional)
nombre_completo('juan','perez',False)
#se pueden ignorar los parametros que ya tienen valores por defecto
nombre_completo('juan','perez')
#llamado de funciones por medio de keywords
nombre_completo(apellido='perez',nombre='juan')
```

```python
#alcance de las funciones y de las variables (scope)
"""
contexto global: cuando se declara una variable global esto significa que su valor y definicion son reconocidas en todo el programa
contexto local: cuando una variable se declara de forma local significa que el valor y definicion solo son reconocidos por la funcion que definio la variable
ejemplo:

"""
def funcion_1(par1, funcion_rec):
    def funcion_2(par2):
        return par2 *2
    valor = funcion_2(par1)
    return funcion_rec(valor)

def funcion_3(par3):
    return par3 + 5
funcion_1(5,funcion_3)
```

```python
#comentareado o especificaciones de codigo (doc strings)
"""para hacer un comentario de varias lineas se hace con triple comillas dobles
partes importantes en un doc strings:

primero: definicion corta de el funcionamiento del codigo o funcion que se desea comentarear
segundo: que significan los parametros (que se espera recibir)
tercero: que es lo que retorna el programa o funcion

es importante generar estos comentarios ya que cuando otro usuario necesite saber que hacen nuestras funciones podra saberlos por medio del comando help()
"""
#para hacer un comentario de una sola linea se hace con numeral (#)

def multiplicacion(a,b):
    """
    funcion que multiplica dos numeros

    recibe dos entero cualquiera
    a = entero cualquiera
    b = entero cualquiera

    return resultado de la multiplicacion entre a y b
    """
    return a * b

multiplicacion(5,5)
help(multiplicacion) # es lo que va a ver el usuario al invocar el comando help
```

```python
#recursividad
"""
de forma algoritmica recursividad se traduce en la capacidad de dividir un problema en problemas mas pequenos para asi poder tener una solucion
mas rapida y eficiente
de forma programatica recursividad es una tecnica mediante la cual una funcion se llama asi misma
"""
def factorial(n):
    """
    calcula el factorial de n

    n entero positivo

    return n!
    """
    if n ==  1:
        return 1
    return n * factorial(n - 1)
n = int(input('ingrese un entero positivo:'))
factorial(n)
#el limite por omision en el numero de iteraciones en la recursividad en python es de 1000 llamadas recursivas, es posible cambiar el tam maximos de la pila
#mediante la instruccion  sys.setrecursionlimit(n)
#cabe recordar que como en python todo es un objeto, las funciones pueden ser enviadas como parametro ya que estas (en python) tambien son objetos
```

```python
#tuplas
#es una secuencia inmutable de objetos
#para declarar una tupla se puede hacer de las siguientes maneras
tup1 = ()#con esto se declara una tupla vacia
tup2 = tuple()
print(type(tup1))
print(type(tup2))
tup1 = (1,'dos',True)
tup1[1]#las tuplas se puede acceder por indices
#al ser inmutables no se pueden modificar los valores
#tup1[1] = 2 # esto no se puede hacer
#las tuplas se pueden re asignar pero no se pueden modificar los valores que contienen
#para crear una tupla de un solo valor
my_tuple = (1)#esto no funciona dado que python lo toma como entero para poder crearlo como tupla necesitamos la coma
print(type(my_tuple))
my_tuple = (1,)
print(type(my_tuple))
#tambien se pueden hacer operaciones entre tuplas (ejemplo sumar dos tuplas)
tup1 = (1,2,3,4)
tup2 = (2,3,4,5)
tup1 += tup2
print(tup1)# no es que se haya modificado tup1 lo que sucedio es que se creo una nueva tupla y le asignamos el mismo nombre tup1 (reasignacion)
#para desempaquetar una tupla en variable
x,y,z,w = tup2
print(x,y,z,w)#es importante tener en cuenta que para aplicar esto debemos desempaquetar todos los elementos
```

```python
#rangos
#representan una secuencia de enteros
#se pueden generar con un comienzo fin y pasos a recorrer
#normamente se utilizan en los for loops
#los rangos tambien son inmutables
my_range = range(1,5)
print(type(my_range))
for i in my_range:
    print(i)
#recordar que los rangos son no inclusivos es decir si uno define un rango entre 1,5 cuando se recorre sera (1,2,3,4)
#prueba de rangos a 2 pasos
range1 = range(0,7,2)
range2 = range(0,8,2)
for i in range1:
    print(i)
print('otro rango')
for i in range2:
    print(i)
#podemos notar que arroja el mismo resultado esto se debe que al no ser inclusivos no toma el limite final por lo tanto no alcanza a llegar a 8 (por lo que es cada 2 pasos)
#para ver que son dos rangos diferentes observemos sus direcciones en memoria
print(id(range1))
print(id(range2))
#esto tambien lo podemos ver por medio de un object equaliti (igualdad entre objetos (is))
print(range1 is range2)
#ejemplo final generar un range con todos los numeros pares desde el 12 hasta el 100
numbers_range = range(0,101,2)
for i in numbers_range:
    print(i)
```

```python
#listas
#ESTA SI ES MUTABLE
#esto en memoria significa que cuando se declara python no sabe que numero de memoria reservar por lo que hay que ser muy cuidadoso con ellas
#al ser mutables, hay que tener varias cosas en cuenta si no queremos tener efectos secundarios (side effects)
#es posible iterar con ellas
#hay que tener mucho cuidado al modificar una lista (esto puede ser una fuente de bugs en el programa)
#formas de declarar una lista
#las listas pueden contener diferentes tipos de datos
my_list = []
my_list2 = list() 
my_list = [1,2,3]
print(my_list[0]) #se pueden acceder a los valores por medio de indices
print(my_list[1:])#tambier se puede utilizar la denotacion de slices
my_list.append(4)#se agrega un valor
print(my_list)
my_list[0] = 'a' #se pueden modificar los valores de las listas
print(my_list)
print(my_list.pop())#retorna el ultimo elemento y a us ves lo eliminina
print(my_list)
for element in my_list:#se puede iterar en una lista
    print(element)
#side effects
lista = [1,2,3]
listb = lista
#en principio creeriamos que tenemos dos variables pero no es asi, lo qeu sucede es que tenemos dos nombres apuntanto a una misma direccion en memoria
print(lista)
print(listb)
print(id(lista))# con id podemos ver que ambas variables tienen la misma direccion
print(id(listb))
c = [lista,listb]
print(c)
lista.append(5)
print(c)
#para no tener este error se debe copiar una lista de esta manera (clonacion)
temp = list(lista)#tambien se puede hacer con slices temp = lista[::]
print(id(lista))
print(id(temp))
#list comprehension
"""es una forma concisa de aplicar operaciones a los valores de una secuencia"""
my_list = list(range(100))
print(my_list)
double = [i*2 for i in my_list]
print(double)
pares = [i for i in my_list if i %2 == 0]
print(pares)
```

```python
#diccionarios (en otros lenguajes conocido como mapa de hash)
"""los diccionarios en luigar de utilizar indices numerico utilizan llaves (keys).
es un "vector de parejas" llave: valor donde las llaves en el diccionario deben ser unicas
los diccionarios no tienen orden interno por los que se manejan por medio de llaves.
los diccionarios son mutables
tambien pueden iterarse  a traves de sus laves, valores y llaves junto con valores
"""
my_dict =  {
    'david': 35,
    'erika': 32,
    'jaime': 50
}
print(my_dict)
print(my_dict['david'])
#print(my_dict['erik']) si queremos acceder a una llave que no existe en el diccionario con esta notacion nos arroja error, para podrer hacerlos debemos 
print(my_dict.get('erik',False))#retorna falso si no encuentra la llave
my_dict['david'] = 10
print(my_dict)
del my_dict['jaime']
print(my_dict)
my_dict['pedro'] = 70
for key in my_dict.keys():#se recorre el diccionario por las llaves
    print(key)
for value in my_dict.values():#se recorre el diccionario por los valores
    print(value)
for llave,valor in my_dict.items():#se recorre el diccionario por llave y valor
    print(llave,valor)
print('david' in my_dict)#tambien sirve para saber si la llave esta o no en el diccionario
```
