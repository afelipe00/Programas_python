<h2>Clustering con Kmeans<h2>

```python
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv('visualClusters.csv')
datos = df.values

print(datos[0:10])
```

```python
xs = datos[:,0]
ys = datos[:,1]
plt.scatter(xs, ys)
plt.show()
```

```python
#Vamos a utilizar Skitlear Kmeans
from sklearn.cluster import KMeans

model = KMeans(n_clusters=3)
model.fit(datos)
labels = model.predict(datos)
print(model)
print(labels)
plt.scatter(xs, ys, c=labels)
plt.show()
```

```python
#Imprimamos ahora los centroides
centroids = model.cluster_centers_
print(centroids)
plt.scatter(xs, ys, c=labels)
plt.scatter(centroids[:,0], centroids[:,1], marker='X', s=200)
plt.show()
```

```python
##Y ahora como predecimos nuesvos puntos
new_df = pd.read_csv('visualClusterPredict.csv')
new_datos = new_df.values
new_labels = model.predict(new_datos)
plt.scatter(new_datos[:,0], new_datos[:,1], c=new_labels)
plt.scatter(centroids[:,0], centroids[:,1], marker='X', s=200)
plt.show()
```

<h2>IRIS dataset vamos a determinar el nmero mas apropiado para k</h2>

```python

iris_df = pd.read_csv('iris.csv')
iris_df
```

```python
#No necesitamos el label por ahora
varieties = list(iris_df['variety'])
del iris_df['variety']
iris_df
```

```python
iris_values = iris_df.values
iris_values[0:10]
```

```python
ks = range(1, 10)
inertias = [] #inertia :Sum of squared distances of samples to their closest cluster center.

for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters=k)

    # Fit model to samples
    model.fit(iris_values)

    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
    
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

<h4>k=3 parece apropiado, fijense que corresponde al nmero de clases que existen en el dataset</h4>

```python
model = KMeans(n_clusters=3)
model.fit(iris_values)
labels_iris = model.fit_predict(iris_values)
df_comparison = pd.DataFrame({'labels': labels_iris, 'varieties': varieties})
df_comparison
```

```python
tabla_pivot = pd.crosstab(df_comparison['labels'], df_comparison['varieties'])
tabla_pivot
```
