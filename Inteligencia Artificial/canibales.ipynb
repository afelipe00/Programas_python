```python
#se definen los lados de las orillas
derecha = -1
izquierda = 1

class Estado:
    #variables privadas privadas que se establecen como iniciales
    _canibales_der = 3
    _misioneros_der = 3
    _canibales_izq = 0
    _misioneros_izq = 0
    _pos_barca = derecha
    _estado_anterior = None
    #se define el metodo constructor
    def __init__(self, canibales_der, misioneros_der, canibales_izq, misioneros_izq, pos_barca):
        self._canibales_der = canibales_der
        self._misioneros_der = misioneros_der
        self._canibales_izq = canibales_izq
        self._misioneros_izq = misioneros_izq
        self._pos_barca = pos_barca
    #metodo para mostrar objetos
    def __str__(self):
        this_barca_der = self._pos_barca == derecha
        return self._canibales_der * " ♠ " + self._misioneros_der * ' ○ '+ this_barca_der * ' ▬▬▬ ' + ' / ' + \
        self._canibales_izq * ' ♠ ' + self._misioneros_izq * ' ○ ' + (not this_barca_der) * ' ▬▬▬ '
    #metodo representativo
    def __repr__(self):
        return self.__str__()
    #metodo de equivalencia
    def __eq__(self, otro):
        return self._misioneros_der == otro._misioneros_der and \
        self._canibales_der == otro._canibales_der and \
        self._misioneros_izq == otro._misioneros_izq and \
        self._canibales_izq == otro._canibales_izq and \
        self._pos_barca == otro._pos_barca
    #metodo que verifica un estado valido
    def estado_valido(self):
        #se verifica que el numero de canibales en cada lado sea el correcto
        if not (0 <= self._canibales_der <= 3 and 0 <= self._canibales_izq <= 3 and 0 <= self._misioneros_der <= 3 and 0 <= self._misioneros_izq <= 3):
            #print('pass')
            return False
        #se verifica que el numero total de canibales y misioneros sea el correcto
        if self._misioneros_der + self._misioneros_izq > 3 or self._canibales_der + self._canibales_izq > 3:
            #print('pass1')
            return False
        #se verifica que no hayan mas canibales que misioneros en la derecha
        if self._misioneros_der > 0 and self._misioneros_der < self._canibales_der:
            #print('pass2')
            return False
        #se verifican que no hayan mas canibales que misioneros en la izquierda
        if self._misioneros_izq > 0 and self._misioneros_izq < self._canibales_izq:
            #print('pass3')
            return False
        return True
```

```python
from copy import copy

def new_estates(actual):
    lst_estados = list()
    signo = actual._pos_barca
    #se va un misionero
    nuevo = copy(actual)
    nuevo._estado_anterior = actual
    nuevo._misioneros_der = nuevo._misioneros_der + signo* 1
    nuevo._misioneros_izq = nuevo._misioneros_izq - signo * 1
    nuevo._pos_barca = -1 * nuevo._pos_barca
    if nuevo.estado_valido():
        #print('se fue un canival')
        lst_estados.append(nuevo)
    #se van dos misioneros
    nuevo = copy(actual)
    nuevo._estado_anterior =  actual
    nuevo._misioneros_der = nuevo._misioneros_der + signo * 2
    nuevo._misioneros_izq = nuevo._misioneros_izq - signo * 2
    nuevo._pos_barca = nuevo._pos_barca * -1
    if nuevo.estado_valido():
        #print('se fue dos canival')
        lst_estados.append(nuevo)
    #se va un canibal
    nuevo = copy(actual)
    nuevo._estado_anterior = actual
    nuevo._canibales_der = nuevo._canibales_der + signo * 1
    nuevo._canibales_izq = nuevo._canibales_izq - signo * 1
    nuevo._pos_barca = -1 * nuevo._pos_barca
    if nuevo.estado_valido():
        #print('se fue dos canival')
        lst_estados.append(nuevo)
    #se van dos canibales
    nuevo = copy(actual)
    nuevo._estado_anterior = actual
    nuevo._canibales_der = nuevo._canibales_der + signo * 2
    nuevo._canibales_izq = nuevo._canibales_izq - signo * 2
    nuevo._pos_barca = -1 * nuevo._pos_barca
    if nuevo.estado_valido():
        #print('se fue dos canival')
        lst_estados.append(nuevo)
    #se va un canibal y un misionero
    nuevo = copy(actual)
    nuevo._estado_anterior = actual
    nuevo._misioneros_der = nuevo._misioneros_der + signo *1
    nuevo._misioneros_izq = nuevo._misioneros_izq - signo * 1
    nuevo._canibales_der = nuevo._canibales_der + signo *1
    nuevo._canibales_izq = nuevo._canibales_izq - signo *1
    nuevo._pos_barca = nuevo._pos_barca * -1
    if nuevo.estado_valido():
        #print('se fue un canival y misonero')
        lst_estados.append(nuevo)
    return lst_estados
```

```python
a = Estado(1,1,2,2,izquierda)
print(a)
nuevos = new_estates(a)
print(len(nuevos))
[x.__str__() for x in nuevos]
```

```python
def sol_out(estado_actual):
    estado = estado_actual
    solucion = []

    while estado is not None:
        solucion.append(estado)
        estado = estado._estado_anterior
    for e in reversed(solucion):
        print(e)
```

```python
#BUSQUEDA POR PROFUNDIDAD
estado_inicial = Estado(3 ,3 ,0 ,0 , derecha)
estado_final = Estado(0 ,0 ,3 ,3 , izquierda)

print('busqueda por profundidad')
estado_actual = estado_inicial
pila = []
visitados = []
visitados.append(estado_actual)
while not estado_actual.__eq__(estado_final):
    Estados_vecinos=new_estates(estado_actual)
    
    pila.extend(Estados_vecinos)
    while True:
        #print(Estados_vecinos)
        estado_actual=pila.pop()
        if estado_actual not in visitados:
            visitados.append(estado_actual)
            break
sol_out(estado_actual)

print("\nbusqueda primero en anchura")

estado_actual = estado_inicial
cola = []
visitados = []
visitados.append(estado_actual)
while not estado_actual.__eq__(estado_final):
  Estados_vecinos=new_estates(estado_actual)
  cola.extend(Estados_vecinos)
  while True:
    estado_actual=cola.pop(0)
    if estado_actual not in visitados:
      visitados.append(estado_actual)
      break
sol_out(estado_actual)
```
